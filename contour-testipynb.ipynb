```python
import nibabel as nib
import numpy as np
import nilearn
import nilearn.masking as masking
import nilearn.plotting as plotting
import matplotlib.pylab as plt
import scipy
import os
```

```python
folder = "Burdenko-GBM-059/RTplanning"
fig_folder = f"{folder}/Figures"
os.makedirs(fig_folder, exist_ok=True)

rtdose_nifti = nib.load(f"{folder}/RTDOSE_Slicer3D.nii.gz")
ct_nifti = nib.load(f"{folder}/CT.nii.gz")
flair_nifti = nib.load(f"{folder}/FLAIRxCT_Warped.nii.gz")
cet1_nifti = nib.load(f"{folder}/CET1xCT_Warped.nii.gz")
```

### RTDOSE aligned to CT

```python
# Around 30 seconds per subject
iso_levels = np.unique(np.int64(rtdose_nifti.get_fdata()))[::5]

fig, ax = plt.subplots(3, 1, figsize=(20, 8), facecolor="k")
fig.subplots_adjust(hspace=0, left=0, right=1, top=1, bottom=0)

display_x = plotting.plot_anat(
    ct_nifti, 
    axes=ax[0], 
    display_mode="x", 
    cut_coords=np.linspace(-80,80,10), 
    draw_cross=False, 
    black_bg=True
)
display_x.add_contours(
    rtdose_nifti,
    levels=iso_levels,
    cmap = 'Reds',
    linewidths=.25,
    antialiased=False
)

display_y = plotting.plot_anat(
    ct_nifti, 
    axes=ax[1], 
    display_mode="y", 
    cut_coords=np.linspace(-120,90,10), 
    draw_cross=False, 
    black_bg=True
)
display_y.add_contours(
    rtdose_nifti,
    levels=iso_levels,
    cmap = 'Reds',
    linewidths=.25,
    antialiased=False
)

display_z = plotting.plot_anat(
    ct_nifti, 
    axes=ax[2], 
    display_mode="z", 
    cut_coords=np.linspace(-120,70,10), 
    draw_cross=False, 
    black_bg=True
)
display_z.add_contours(
    rtdose_nifti,
    levels=iso_levels,
    cmap = 'Reds',
    linewidths=.25,
    antialiased=False
)

fig.suptitle("Burdenko-GBM-059", fontweight='bold', color='white', fontsize=20)
fig.savefig(f"{fig_folder}/RTDOSE_registration.svg", format='svg', dpi=300)
fig.savefig(f"{fig_folder}/RTDOSE_registration.png", format='png', dpi=300)
fig.savefig(f"{fig_folder}/RTDOSE_registration.pdf", format='pdf', dpi=300)
```
